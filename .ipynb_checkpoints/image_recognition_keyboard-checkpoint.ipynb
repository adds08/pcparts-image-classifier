```python
from PIL import Image
import numpy as np
import os
import matplotlib.pyplot as plt

# Define a function to load images from a directory
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = Image.open(os.path.join(folder, filename))
        if img is not None:
            images.append(np.array(img))
    return images

# Provide the path to the directory containing images
folder_path = '/content/drive/MyDrive/Datasets/pc_parts/keyboard'
foldernot_path = '/content/drive/MyDrive/Datasets/pc_parts/monitor'

# Load images from the specified directory
images_keyboard = load_images_from_folder(folder_path)
images_monitor = load_images_from_folder(foldernot_path)

# Convert the list of images to a NumPy array
images_np = np.array(images_keyboard + images_monitor)

# Check the shape of the NumPy array
print("Shape of the NumPy array containing images:", images_np.shape)


```

```python
show_image_count = 0
fig, axes = plt.subplots(1, 4, figsize=(12, 3))

for filename in os.listdir(folder_path):
  img = Image.open(os.path.join(folder_path, filename))
  if img is not None:
    if(show_image_count>=4):
      break
    ax = axes[show_image_count]
    ax.imshow(img)
    ax.axis("off")
  show_image_count+=1

plt.tight_layout()
plt.show()
```

```python
show_image_count = 0
fig, axes = plt.subplots(1, 4, figsize=(12, 3))

for filename in os.listdir(foldernot_path):
  img = Image.open(os.path.join(foldernot_path, filename))
  if img is not None:
    if(show_image_count>=4):
      break
    ax = axes[show_image_count]
    ax.imshow(img)
    ax.axis("off")
  show_image_count+=1

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

labelsKeyboard = [1] * len(images_keyboard)  # Assigning label 1 for keyboards
labelsMonitor = [0] * len(images_monitor)  # Assigning label 1 for keyboards
labels_np = np.array(labelsKeyboard+labelsMonitor)

#split data into test and test
X_train, X_val, y_train, y_val = train_test_split(images_np,labels_np, test_size=0.2, random_state=42)

# Normalize pixel values to be between 0 and 1
train_images = X_train / 255.0
test_images = y_train / 255.0
```

```python
import tensorflow as tf
from keras import layers, models, utils

#Define CNN
model = models.Sequential([
    layers.Conv2D(64, (5, 5), activation='relu', input_shape=(256, 256, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (5, 5), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (5, 5), activation='relu'),
    layers.Flatten(),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')  # Binary classification (keyboard or not)
])

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the model
model.fit(train_images,test_images, epochs=10, batch_size=32, validation_data=(X_val,y_val))

```

```python
# Evaluate the model
test_loss, test_acc = model.evaluate(X_val,y_val)
print('Test accuracy:', test_acc)
```
